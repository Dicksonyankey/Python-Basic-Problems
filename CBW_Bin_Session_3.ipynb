```python
class Book(object):
    def __init__(self, title, author, isbn, is_available):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_available = is_available

    def get_details(self):
        print("Book Details")
        return f"Title : {self.title}\nAuthor : {self.author}\nISBN : {self.isbn}\nAvailable : {self.is_available}"

    def __str__(self):
        return f"Book({self.title}, {self.author}, {self.isbn}, {self.is_available})"


class Patron(object):
    def __init__(self, name, id, no_of_books_borrowed = 0):
        self.name = name
        self.id = id
        self.no_of_books_borrowed = no_of_books_borrowed

    def get_details(self):
        print("Patron Details")
        return f"Name : {self.name}\nID : {self.id}\nNumber of books borrowed : {self.no_of_books_borrowed}"

    def __str__(self):
        return f"patron({self.name}, {self.id}, {self.no_of_books_borrowed})"

class LibraryCatalog(object):
    def __init__(self):
        self.books = []
        self.patrons = []
        self.tracking_books = {}

    def add_book(self, book: Book):
        if book and isinstance(book, Book):
            self.books.append(book)
            print(f"Book '{book.title}' by {book.author} added to the catalog.")
        else:
            print("Invalid Book Object")
        self.display_books()

    def remove_book(self, isbn: str):
        self.books = list(filter(lambda x : x.isbn != isbn, self.books))
        print(f"Book with isbn number {isbn} has been removed.")
        self.display_books()

    def add_patron(self, patron: Patron):
        if patron and isinstance(patron, Patron):
            self.patrons.append(patron)
            print(f"{patron.name} has been added to the library system.")
        else:
            print("Invalid Patron Object")
        self.display_patrons()


    def remove_patron(self, patron_id: int):
        self.patrons = list(filter(lambda x : x.id != patron_id, self.patrons))
        print(f"Patron with isbn number {patron_id} has been removed.")
        self.display_patrons()

    def borrow_book(self, patron_id: int, isbn: str):
        """allow a patron to borrow a book. Update book availability andpatron's borrowed books."""
        # next() function with a generator expression to find the patron and book objects.
        filtered_patron = next((patron for patron in self.patrons if patron.id == patron_id), None)
        filtered_book = next((book for book in self.books if book.isbn == isbn), None)
        if filtered_patron and filtered_book and filtered_book.is_available:
            filtered_patron.no_of_books_borrowed += 1
            filtered_book.is_available = False
            print(f"Book '{filtered_book.title}' has been borrowed by patron ID {patron_id}.")
            print(f"Total books borrowed by patron ID {patron_id}: {filtered_patron.no_of_books_borrowed}")
            print(f"Availability of '{filtered_book.title}': {filtered_book.is_available}")
        elif not filtered_patron:
            print(f"No patron found with ID {patron_id}.")
        elif not filtered_book:
            print(f"No book found with ISBN {isbn}.")
        else:
            print(f"Book '{filtered_book.title}' is not available for borrowing.")
        self.display_patrons()
        self.display_books()


    def return_book(self, patron_id: int, isbn: str):
        pass

    def get_borrowed_books(self, patron_id: int):
        pass

    def display_books(self):
        print("Books in the catalog:")
        for book in self.books:
            print(book)
    def display_patrons(self):
        print("Books in the catalog:")
        for patron in self.patrons:
            print(patron)





# Creating a Book
book1 = Book("The Catcher in the Rye", "J.D. Salinger", "9780316769488", True)
book2 = Book("To Kill a Mockingbird", "Harper Lee", "9780061120084", True)
print(book1)
print(book1.get_details())

print("\n----------------------------")
# Creating Patrons
patron1 = Patron("Alice Johnson", 1)
patron2 = Patron("Bob Smith", 2)
print(patron1)
print(patron1.get_details())

print("\n----------------------------")
# Creating Library Catalog
library = LibraryCatalog()

# Adding and Removing book from the book Catalog.
library.add_book(book1)
# library.remove_book("9780316769488")
# library.books

# Adding and Removing Patron from library system
library.add_patron(patron1)
library.add_patron(patron2)

print("\n----------------------------")
library.borrow_book(1, "9780316769488")
```
